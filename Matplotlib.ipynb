```python
from matplotlib import pyplot as plt
import numpy as np
x_values = [1,2,3,4]
y_values = [5,4,6,2]
plt.scatter(x_values,y_values)

other_x_values = [1,2,3,4]
other_y_values = [1,2,3,4]
plt.plot(other_x_values,other_y_values, color = "red")

plt.title("Scatter Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,5,11)
y = x**2
x
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(x,y)
plt.title("Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.subplot(1,2,1) #row,coloum,(r or c any use)
plt.plot(x,y,color = "red")
plt.title("Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.subplot(1,2,2)
plt.plot(y,x,color = "green")
plt.title("Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height(range 0 to 1)
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y,'r')
axes1.set_xlabel("X Label")
axes1.set_ylabel("Y Label")
axes1.set_title("Set Title01")

axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x,'b')
axes2.set_xlabel("X Label")
axes2.set_ylabel("Y Label")
axes2.set_title("Set Title02")
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y,color = "red")
axes[0].plot(y,x,color = "red")

axes[0].plot(x,y,color ="blue")
axes[1].plot(y,x,color="blue")


type(axes)
```

```python
fig = plt.figure(figsize = (8,2), dpi = 80) #  width, height
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Set Title")
```
